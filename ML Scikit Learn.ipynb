**IMPORTING LIBRARY (SCI-KIT LEARN)**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
x1 = [7,6,4,3]
x2 = [5,4,7,8]
target = ['bad','bad','good','good']
feature = list(zip(x1,x2))

feature
```

**MODEL BUILDING**

```python
model = KNeighborsClassifier(n_neighbors = 3)
```

**MODEL TRAINING**

```python
model.fit(feature,target)
```

**MODEL PREDICT**

```python
model.predict([[3,7]])
```
